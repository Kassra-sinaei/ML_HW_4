#### import libraries

```python
import numpy as np
from matplotlib import pyplot as plt
```

#### calculate distance (Euclidean )

```python
np.random.seed(1)

def Euc_dist(x, y):
    return np.linalg.norm(x - y)

def do_task(dim, size):
    # Generate points
    points = list("")
    for i in range(size):
        points.append(np.random.rand(dim))
    # Calculate Distances
    dist = list("")
    for i in range(len(points)):
        for j in range(i + 1, len(points)):
            dist.append(Euc_dist(points[i], points[j]))

    # Plot Histogram of Distances
    fig, ax = plt.subplots()
    fig.set_dpi(100)
    n, bins, patches = ax.hist(dist, 10, density=True, facecolor='g', alpha=0.75)
    ax.set_xlabel("Euclidean Distance")
    ax.set_ylabel("Frequency")
    ax.set_title("Dimention = " + str(dim) + ", Size = " + str(size), fontsize=20)
    plt.grid(True)
    plt.show()
```

#### Part A

```python
do_task(5, 5)
do_task(10, 5)
do_task(100, 5)
do_task(500, 5)
do_task(1000, 5)
do_task(10**4, 5)
do_task(10**6, 5)
```

#### Part B

```python
do_task(10**4, 5)
do_task(10**4, 10)
do_task(10**4, 100)
do_task(10**4, 5000)
do_task(10**4, 10**4)
```
